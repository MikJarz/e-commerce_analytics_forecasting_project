```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading and preparing data

```python
# Load datasets
orders = pd.read_csv('orders.csv')
products = pd.read_csv('products.csv')
sellers = pd.read_csv('sellers.csv', delimiter=',')
order_items = pd.read_csv('order_items.csv')
customers = pd.read_csv('customers.csv')
```

```python
# Convert timestamps to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
# Convert missing values to mean
products['product_weight_g'].fillna(products['product_weight_g'].mean(), inplace=True)

# Merge dataframes
merged_df = pd.merge(order_items, products, on='product_id')
merged_df = pd.merge(merged_df, sellers, on='seller_id')
merged_df = pd.merge(merged_df, orders, on='order_id')
merged_df = pd.merge(merged_df, customers, on='customer_id')
```

```python
order_items
```

```python
# Calculate the total price (price for products + shipping cost a.k.a. freight value)
merged_df['total_price'] = merged_df['price'] + merged_df['freight_value']

# Calculate the turnover for each category
category_turnover = merged_df.groupby('product_category_name')['total_price'].sum().sort_values()
```

# Biggest/smallest turnover

```python
# Calculate seller turnover
seller_turnover = merged_df.groupby('seller_id')['total_price'].sum().sort_values()
```

```python
seller_turnover
```

```python
seller_turnover.describe()
```

```python
# Calculate top and bottom sellers
top_sellers = seller_turnover.nlargest(5)
bottom_sellers = seller_turnover.nsmallest(5)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plotting top sellers
axes[0].bar(top_sellers.index, top_sellers.values, color='blue')
axes[0].set_title('Top 5 Sellers')
axes[0].set_xlabel('Seller ID')
axes[0].set_ylabel('Turnover')
axes[0].tick_params(axis='x', rotation=90)

# Plotting bottom sellers
axes[1].bar(bottom_sellers.index, bottom_sellers.values, color='red')
axes[1].set_title('Bottom 5 Sellers')
axes[1].set_xlabel('Seller ID')
axes[1].set_ylabel('Turnover')
axes[1].tick_params(axis='x', rotation=90)

plt.show()
```

# Leaders/outsiders in sales in each area

```python
# Calculate total sales by state
state_sales = merged_df.groupby('customer_state')['total_price'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(15, 8))
state_sales.plot(kind='bar', color='purple')
plt.title('Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.show()
```

# Turnover by category

```python
# Calculate turnover by product category
category_turnover = merged_df.groupby('product_category_name')['total_price'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(15, 8))
category_turnover.plot(kind='bar')
plt.title('Turnover by Product Category')
plt.xlabel('Category')
plt.ylabel('Turnover')
plt.show()
```

# Dependence of product weight on turnover and price (relationship)

```python
plt.figure(figsize=(15, 7))
sns.scatterplot(x='product_weight_g', y='total_price', data=merged_df, alpha=0.5)
plt.title('Product Weight vs Turnover')
plt.xlabel('Product Weight (grams)')
plt.ylabel('Turnover')
plt.grid(True)
plt.show()
```

```python
# Dataframe without extreme outliers (deleted 6 records in total)
df_outliers = merged_df[(merged_df['product_weight_g'] <= 32500) & (merged_df['total_price'] <= 5000)]
```

```python
# Pearson correlation coefficient
pearson_corr = df_outliers['product_weight_g'].corr(df_outliers['total_price'])
pearson_corr
```

```python
plt.figure(figsize=(15, 7))
sns.scatterplot(x='product_weight_g', y='total_price', data=df_outliers, alpha=0.5)
# Plot the regression line based on the correlation coefficient
plt.plot(df_outliers['product_weight_g'], pearson_corr * df_outliers['product_weight_g'], color='red', linestyle='--', alpha=0.5)
plt.title('Product Weight vs Turnover')
plt.xlabel('Product Weight (grams)')
plt.ylabel('Turnover')
plt.grid(True)
plt.show()
```

# Conclusion

Segmentation of Sellers:
*   <b>Low Sales:</b> Sellers with total sales volume below the 25th percentile (e.g., less than 270\\$) can be classified as low performers.
*   <b>Moderate Sales:</b> Sellers with total sales volume between the 25th and 75th percentiles (e.g., 270\\$ to 3,965\\$) can be considered moderate performers.
*   <b>High Sales:</b> Sellers with total sales volume above the 75th percentile (e.g., more than 3,965\\$) are high performers.

Textual conclusions:
*   <b>Low Performers:</b> These sellers may benefit from targeted marketing efforts or product diversification to increase sales and improve their business value. Identifying the reasons behind their low sales can help to fit strategies for improvement.
*   <b>Moderate Performers:</b> These sellers are performing decently but might have room for growth. Strategies to extend their product offerings, optimize pricing, or improve customer service can help unlock their full potential and increase their business value.
*   <b>High Performers:</b> These sellers are driving significant sales volume, indicating strong customer demand for their products. Investing in maintaining customer satisfaction, expanding product lines, and exploring new market opportunities can further capitalize on their success and maximize business value.

<b>Turnover by Product category:</b> Analyzing turnover by product category reveals best-sellers and underperformers, guiding inventory and marketing strategies. It helps optimize stock levels, tailor promotions, spot seasonal trends, and inform pricing and product development. These insights enhance profitability and customer satisfaction.
    
<b>Turnover/weight correlation:</b> Examining the connection between product weight and sales can guide

Additionally, insights from this analysis could help in optimizing packaging solutions or negotiating better shipping rates, further improving profitability for these items.
